## <mark>Exercise #5
**Count Movies by Genre** (Total: 50 points)

- Complete a function `count_movies_by_genre(file, genre)` that reads a CSV file and counts the number of movies in a specific genre.
    - The CSV contains a column labeled `genre`.
    - The function should return the total count of movies that belong to the specified genre.

- Example:
    - Calling `count_movies_by_genre('n_movies.csv', 'action')` should return `100`, meaning there are 100 movies that belong to the genre `'action'`.



```python
# n_movies.csv is a dataset from Kaggle: https://www.kaggle.com/datasets/narayan63/netflix-popular-movies-dataset

import csv

def count_movies_by_genre(file, genre):
    # YOUR CODE HERE
    c = 0
    genre = genre.capitalize()
    with open(file,mode = 'r',newline='', encoding='cp1252') as mycsv:
        reader = csv.DictReader(mycsv)
        for line in reader:
            colum_genre = line["genre"].split(", ")
            if genre in colum_genre:
                c += 1
        return c
```

## <mark>Exercise #6
**Count Movies by Genre and Rating from CSV** (Total: 40 points)</mark>

- Complete a function `read_csv()` to process a CSV file containing movie data.
    - Each movie entry should include details like **title**, **genre**, **rating**, and more.
    - The function should accept the following parameters:
        1) `file`: The path to the CSV file.
        2) `genre`: The specific movie genre to filter by.
        3) `rating`: The minimum movie rating to consider.
    - The function should return three values:
        1) The total number of movies that belong to the given genre and have a rating **greater than or equal to** the specified `rating`.
        2) The title of the **highest-rated** movie from the filtered results.
        3) The title of the **lowest-rated** movie from the filtered results.
    
    - Example:
        - If you call `read_csv('n_movies.csv', 'Fantasy', 7.5)`, the function should return:
            - `(185, 'Stranger Things', 'All of Us Are Dead')`
            - Here, `185` represents the number of movies in the `Fantasy` genre with a rating of **7.5 or higher**.
            - `'Stranger Things'` is the title of the movie with the **highest rating** among those 185 movies.
            - `'All of Us Are Dead'` is the title of the movie with the **lowest rating** among those 185 movies.



```python
import csv

def read_csv(file, genre, rating):
    # YOUR CODE HERE
    genre = genre.capitalize()
    max = rating
    min = rating
    c = 0
    with open(file,mode = 'r',encoding='cp1252') as mycsv:
        reader = csv.DictReader(mycsv)
        for colum in reader:
            if genre in colum["genre"]:
                if colum['rating']:
                    if rating <= float(colum['rating']):
                        c += 1
                        if float(colum['rating']) >= max:
                            max = float(colum['rating'])
                            title_max = colum['title']
                        if float(colum['rating']) <= min:
                            min = float(colum['rating'])
                            title_min = colum['title']
        return c,title_max,title_min
```
