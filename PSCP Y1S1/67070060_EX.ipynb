สร้างฟังก์ชัน `create_2d_array(n, m, step)` ที่รับพารามิเตอร์สามตัว:

* `n`: จำนวนแถว
* `m`: จำนวนคอลัมน์
* `step`: ค่า Step ระหว่างองค์ประกอบในอาร์เรย์

ฟังก์ชันควร return ค่าเป็นอาร์เรย์ NumPy แบบ **2-D** โดยแต่ละแถวจะมีตัวเลขเริ่มต้นจาก 0 และมีระยะห่างเท่ากับ `step`

ตัวอย่าง:
- ถ้าให้ค่า `n` เป็น 5 ฟังก์ชันจะคืนค่าเป็น **NumPy** อาร์เรย์ `[ 0,  2,  4],[ 6,  8, 10]`

*หมายเหตุ ใน Python เลข `1` เท่ากับ `1.0`*

```python
def create_2d_array(n, m, step):
    # YOUR CODE HERE
    arr1 = np.arange(0,n*m*step,step)
    return arr1.reshape(n,m)
```

เขียนฟังก์ชัน `normalize_scores(scores)` สำหรับ **ปรับมาตรฐานคะแนน** ของนักเรียนในชั้นเรียน ซึ่งจำเป็นต้องถูกปรับมาตรฐานเพื่อให้ค่าทั้งหมดอยู่ในช่วง 0 ถึง 1
- ฟังก์ชัน NumPy รับอาร์เรย์ **2-D** ซึ่งแต่ละแถวแทนคะแนนของนักเรียนในวิชาต่าง ๆ และ return ค่าเป็นอาร์เรย์ **2-D** ที่มีคะแนนที่ถูกปรับมาตรฐานแล้ว
- สูตรสำหรับการปรับมาตรฐานคะแนน คือ `ค่าที่ปรับมาตรฐาน = (ค่า - ค่าต่ำสุด) / (ค่าสูงสุด - ค่าต่ำสุด)`

ตัวอย่าง
- หาก `scores = np.array([[20, 75, 100], [20, 40, 60], [10, 20, 30]])`
- `normalized = normalize_scores(scores)` จะให้ผลลัพธ์เป็น  `[[0. , 0.5, 1. ],[0. , 0.5, 1. ],[0. , 0.5, 1. ]]`

```python
def normalize_scores(scores):
    # YOUR CODE HERE
    return np.divide(scores - scores.min(axis=1).reshape(2,1), (scores.max(axis=1)-scores.min(axis=1)).reshape(2,1))
    
```

<mark>DO NOT CHANGE THE FOLLOWING LINES</mark>

<mark>**Some tests are hidden**</mark>

```python
# Test 1: 2 points
scores = np.array([[1000, 2000, 3000], [4000, 5000, 6000]])
expected = np.array([[0., 0.5, 1.], [0., 0.5, 1.]])
result = normalize_scores(scores)
assert_array_almost_equal(result, expected)
```

```python
# Test 2: 2 points
scores = np.array([[90, 85, 95], [10, 5, 15]])
expected = np.array([[0.5, 0., 1.], [0.5, 0., 1.]])
result = normalize_scores(scores)
assert_array_almost_equal(result, expected)
```

- ฟังก์ชันคืนค่า: `a=0, b=9, c=4, d=5`
- การใช้ `puzzle[a:b, c:d]` จะดึง `puzzle[0:9, 4:5]` ซึ่งส่งผลให้ได้ `'ASTRONAUT'`

```python
def find_index(puzzle, key_word):
    # YOUR CODE HERE
    for i in range(10):
        if key_word in ''.join(puzzle[i:i+1,:].flatten()):#horizontal
            a = i
            b = i+1
            for row in range(10):
                if key_word == ''.join(puzzle[a:b,row:len(key_word)+row].flatten()):
                    c = row
                    d = row+len(key_word)
        elif key_word in ''.join(puzzle[i:i+1,:].flatten()):#vertical
            c = i
            d = i+1
            for colum in range(10):
                if key_word == ''.join(puzzle[colum:len(key_word)+colum,c,d].flatten()):
                    c = row
                    d = row+len(key_word)
    return a,b,c,d
```
